```python
import folium
import pandas as pd
import geopandas as gpd
import numpy as np
```

```python
migun = gpd.read_file(filename='migunNeighbors.geojson')
migun.head(6)
```

```python
migun.columns
```

```python
from folium import DivIcon

m =folium.Map(location=(migun['geometry'].get_coordinates().mean()['y'],migun['geometry'].get_coordinates().mean()['x']),zoom_control=False,zoom_start=14)

folium.GeoJson(migun ,style_function=lambda feature: {
        "color": "black",
        "fillColor": "#009933"
        if feature['properties']['pop_not_protected_prcnt'] <25
        else '#ffff66'if feature['properties']['pop_not_protected_prcnt'] <50
        else '#ffa31a' if feature['properties']['pop_not_protected_prcnt'] <75
        else ' #ff1a1a',
        "weight": 2,
        "opacity":0.8,
        "fillOpacity": 0.6
    }).add_to(m)


for inst in migun['OBJECTID']:
    folium.map.Marker(
    [migun[migun['OBJECTID']==inst]['geometry'].centroid.get_coordinates()['y'], migun[migun['OBJECTID']==inst]['geometry'].centroid.get_coordinates()['x']-0.003],
    icon=DivIcon(
        icon_size=(150,36),
        icon_anchor=(0,0),
        html='<div style="font-weight:bold;font-family:tahoma;font-size: 7t">%s</div>' % str(migun[migun['OBJECTID']==inst]['FNAME'].values[0].replace(',','</br>')),
        )
    ).add_to(m)
m
```

```python
m.save('map_migun.html')
```

```python
import io 
from PIL import Image
img_data = m._to_png(5)
img = Image.open(io.BytesIO(img_data))
img.save('migun-ashkelon.png')
```
