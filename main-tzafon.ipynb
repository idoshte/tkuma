```python
import folium
import pandas as pd
import geopandas as gpd
import numpy as np
```

```python
instructions = gpd.read_file(filename='instructions.geojson')
instructions.head(6)
instructions = instructions [instructions['merchav'] == 'קו העימות']
```

```python
setls = gpd.read_file(filename='kavHaimutSetles.geojson')
setls.head()
```

```python
from folium import DivIcon

m =folium.Map(location=(setls['geometry'].get_coordinates().mean()['y']-0.005,setls['geometry'].get_coordinates().mean()['x']+0.1),zoom_control=False,zoom_start=11)

folium.GeoJson(instructions,  style_function=lambda feature: {
        "color": "black",
        "fillColor":"#CCB8FC",
        "weight": 2,
        "opacity":0.8,
        "fillOpacity": 0.6
    }).add_to(m)

folium.GeoJson(setls,  style_function=lambda feature: {
        "fillColor": "#2DF77A",
        "color": "black",
        "weight": 1,
        "opacity":0.9,
        "fillOpacity":0.7
    }).add_to(m)

for inst in setls['OBJECTID']:
    folium.map.Marker(
    [setls[setls['OBJECTID']==inst]['geometry'].centroid.get_coordinates()['y'], setls[setls['OBJECTID']==inst]['geometry'].centroid.get_coordinates()['x']],
    icon=DivIcon(
        icon_size=(150,36),
        icon_anchor=(0,0),
        html='<div style="font-weight:bold;font-family:tahoma;font-size: 5pt">%s</div>' % str(setls[setls['OBJECTID']==inst]['SETL_NAME'].values[0]),
        )
    ).add_to(m)

# folium.GeoJson(
#     setl,
#     name="Subway Stations",
#     marker=folium.Circle(radius=4, fill_color="orange", fill_opacity=0.4, color="black", weight=1)
# ).add_to(m)   

# #     folium.GeoJson(setl,  style_function=lambda feature: {

# #         "fillColor": "gray",
# #         "color": "black",
# #         "weight": 0.5,
# #         "opacity":0.7,
# #         "fillOpacity":0.2
# #   }).add_to(m)
# for inst in setl['SETL_CODE'].values:
#     folium.map.Marker(
#     [setl[setl['SETL_CODE']==inst]['geometry'].centroid.get_coordinates()['y'], setl[setl['SETL_CODE']==inst]['geometry'].centroid.get_coordinates()['x']],
#     icon=DivIcon(
#         icon_size=(150,36),
#         icon_anchor=(0,0),
#         html='<div style="color:;font-weight:bold;font-family:tahoma;font-size: 6pt">%s</div>' % str(setl[setl['SETL_CODE']==inst]['SETL_NAME'].values[0]),
#         )
#     ).add_to(m)
m
```
